# What this simulation is doing

This notebook runs a **2D compressible Euler** model (`CompressibleFluid2D`) in a periodic box.

- **State variables**: `rho` (density), `u` (x-velocity), `v` (y-velocity), `p` (pressure).
- **Physics**: conservative compressible flow with nonlinear advection and pressure-wave propagation.
- **Numerics**: finite-volume style update with a local Lax–Friedrichs (Rusanov) flux and adaptive CFL time stepping.

## How to read the output

- `rho` and `p` show compressive/expansive wave patterns.
- `u, v` show the velocity field transporting and distorting those patterns.
- As time evolves, smooth initial perturbations can steepen into sharper fronts due to nonlinearity.

So this is the compressible counterpart to the incompressible flow examples.

```python
from autosim.simulations import CompressibleFluid2D as Sim

sim = Sim(
    return_timeseries=True,
    log_level="warning",
    n=64,
    T=0.8,
    dt_save=0.01,
    cfl=0.35,
    parameters_range={
        "gamma": (1.4, 1.4),
        "amp": (0.08, 0.12),
    },
)

data = sim.forward_samples_spatiotemporal(n_samples=2, random_seed=42)
```
