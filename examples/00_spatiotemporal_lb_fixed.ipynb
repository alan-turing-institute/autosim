# What this simulation is doing

This notebook runs a **2D Lattice Boltzmann (D2Q9)** channel-flow model (`LatticeBoltzmann`).

- **State variables shown**: `u`, `v`, `rho`, `vorticity`.
- **Physics setup**: parabolic inflow in a channel, optional cylinder obstacle, no-slip walls, outlet boundary.
- **Dynamics control**:
  - `use_cylinder=True` gives wake shedding behind the obstacle.
  - `use_cylinder=False` with `oscillatory_inlet=True` gives unsteady shear-driven dynamics without an obstacle.

## How to read the output

- `vorticity` is usually the clearest diagnostic for coherent structures and shedding.
- `rho` varies weakly (near-incompressible regime), while `u,v` carry the main flow pattern.

So this notebook is best for intuitive channel-flow / wake dynamics with boundary effects.

```python
from autosim.simulations import LatticeBoltzmann as Sim

# Flow past a cylinder with parabolic inflow (LBM)
sim = Sim(
    return_timeseries=True,
    log_level="warning",
    width=192,
    height=64,
    T=8.0,
    parameters_range={
        "viscosity": (0.015, 0.025),
        "u_in": (0.08, 0.12),
    },
    use_cylinder=False,
    oscillatory_inlet=True
)

data = sim.forward_samples_spatiotemporal(n=1, random_seed=42)
```
